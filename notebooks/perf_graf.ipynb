```python
PERF_INITIAL_MEASUEMENTS = [1.515, 100, 100, 0.941, 100, 48, 58, 1.684, 0.017, 2.735]
PERF_OPT_MEASUEMENTS = [1.482, 4.279, 4.655, 0.451, 100, 0.93, 0.023, 0.499, 0.017, 1.232]
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10']
x = np.arange(len(labels)) 
width = 0.4

plt.rcParams['figure.figsize'] = (10, 6)  # Set the figure size
plt.rcParams['xtick.labelsize'] = 8  # Set the font size of x-axis tick labels
plt.rcParams['ytick.labelsize'] = 8  # Set the font size of y-axis tick labels

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, PERF_INITIAL_MEASUEMENTS, width, label='No Opt')
rects2 = ax.bar(x + width/2, PERF_OPT_MEASUEMENTS, width, label='Opt')

ax.set_xlabel('Questions')
ax.set_ylabel('Values')
ax.set_title('Comparison of optimized and non optimized query execution time')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
